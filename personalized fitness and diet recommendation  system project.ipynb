# feature construction

```python
df_encoded['BMI'] = df_encoded['weight'] / (df_encoded['height'] / 100) ** 2
```

```python
df_encoded
```

# model building

```python
from sklearn.cluster import KMeans

# Number of clusters
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df_encoded['cluster'] = kmeans.fit_predict(df_encoded)

# Inspect cluster centers
print(kmeans.cluster_centers_)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# Initialize an empty list to store the sum of squared distances (inertia) for each K
inertia = []
K_range = range(1, 6)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_encoded)
    inertia.append(kmeans.inertia_)
# Plot the Elbow Method visualization
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o', linestyle='--', color='b')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Sum of Squared Distances)')
plt.xticks(K_range)
plt.grid(True)
plt.show()
```
